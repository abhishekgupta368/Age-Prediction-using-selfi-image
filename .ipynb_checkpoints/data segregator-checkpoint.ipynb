```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import tensorflow as tf
import keras
import cv2
```

```python
class DataSegregator():
    def __init__(self,f_name,folder):
        self.data = pd.read_csv(f_name)
        self.data_by_age = self.data[['file','urls','age']]
        self.folder = folder
    
    def get_test_data(self):
        test_size = int(len(self.data_by_age)*(0.20))
        self.data_by_age = self.data_by_age[:test_size]
        
    def write_folder(self,f):
        if(os.path.isdir(f )==False):
            os.makedirs(f)
            print(str(f)+" file written")
        else:
            print(str(f)+" already exists")
            
    def process_name(self,name):
        text = name.split('/')
        return text[-1]
    
    def write_data(self):
        self.write_folder(self.folder)
        label_lst =  np.unique(self.data_by_age['age'].values)
        label_url = []
        for label in label_lst:
            url =  self.folder+"/"+label
            self.write_folder(url) 
        for x in label_lst:
            dataset =  self.data_by_age[self.data_by_age['age']==x]['file']
            url = self.folder+"/"+x
            print(x)
            c=0
            for data in dataset:
                img_loc= self.process_name(data)
                save_file = url+"/"+img_loc
                try:
                    print("Writing file "+str(img_loc))
                    cv2.imwrite(save_file,cv2.imread(data))
                except Exception as e:
                    print(str(save_file)+" has error in format")
        
```

```python
if __name__ == "__main__":
    dtseg = DataSegregator("new_fairface_train.csv","test_set")
    dtseg.get_test_data()
    dtseg.write_data()
```

```python
dtseg = DataSegregator("new_fairface_val.csv","val_set")
dtseg.write_data()
```
